## A Brief Introduction to Phase Space 
There are two dynamical properties of a mass (point particle) in which physicists take interest, position and velocity. Phase space is an abstract space where the position of particle acts as abscissa and velocities are plotted as ordinate. Thus, phase space allows us to see the characteristics (the various states it can go in) of a particular system. It is particularly 'interesting' to look at the system and analyse it in phase space when it is sometimes difficult to just take a look at the position profile with respect to time. [Phase portraits](https://en.wikipedia.org/wiki/Phase_portrait) are essentially, the drawings in which you can see how the system will behave or have behaved under different initial conditions. If you take a note of every point, a particle has visited with this and that velocity, you can also draw a phase portrait and you'll get to know what could have been the forces behind it, what it had done at various points without even noting down the time.  

For example, the solutions of chaotic systems are generally read in phase space due to their ease to recognise characteristic structures. The analysis of the Phase Space of a system is also important for a statistically large system and have lead to many interesting results from it. In fact, the foundations of Classical Statistical Mechanics are entirely devised on phase space.

## Autonomous System of equations

Heuristically, you can show yourself that any coupled first-order linear differential equation could be written in the form \begin{equation} \frac{dy}{dt} = f(x,y) \\  \frac{dx}{dt} = g(x,y)\end{equation} The system is called <b>autonomous</b> when $ f(x,y)\ \&\ g(x,y) $ are implicit function of time (t). In this article we will be considering these types of systems. The abundance of linear systems in physics comes from the fact that the solution of linear systems follows the superposition principle and that is why we are interested in these types of problems. 

## Linear System
A vector differential equation is what defines the evolution of your <b>dynamical system</b>. One should ask why we are so adamant on solving these equations? So the answer comes from Continuous Dynamical Systems Theory which deals with the differential equations to "solve a system". By solving, we mean, to find every set of real numbers (positions and momenta at each time interval) which will define its state in a space. We usually get this set by integrating the differential equations in some sense. But again it's oversimplification and you would like to read more about [Dynamical systems theory](https://en.wikipedia.org/wiki/Dynamical_systems_theory).

If things were not hard for you, there's another piece of information. Introducing you to a vector linear differential equation of the form <b> X'= AX </b> which contains time derivative of the vector, denoted by <b>X'</b> and is related to vector $X$ through operator matrix A.

Since, in an abstract sense, a vector is an ordered list of real numbers. And a solution of the coupled first linear equations is also a set of numbers. So, it turns out you can naturally write equations of autonomous system in their vector form, which is discussed in the next section.  

### Critical Points 


Any point $(x_0,y_0)$ is called 'critical point' when $f(x_0,y_0) = g(x_0,y_0) = 0$, which implies that the higher derivatives of vector $X$ are also zero at that point. Interested readers are encouraged to prove this unproven fact about higher derivatives being zero (differential geometry can be employed to prove this). They are also called **Equilibrium Points**, as the particle doesn't change its state under the potential/dynamics, it is in. Whenever a Physicist needs an example, she/he goes to a pendulum. So, there are two equilibrium states (a unique $(x,p)$ set) for a simple pendulum under gravity. One is hanging still in the down position and the other is held up straight immobile. These points are *important* as they are unique in their own right. They are unaltered in time, even under gravity (or any external forces). Due to this character, they are special in phase space analysis as they lie on the abscissa (hence a critical point).  

 Near critical points, the autonomous equation can take the linear form. To see that, let's perturb our particle from this $equilibrium\ position$. We can write this mathematically through following expressions: \begin{equation} x(t) = x_0 + \delta x \\ y(t) = y_0 + \delta y\end{equation}

Also the perturbation should be small (that's why they are perturbations) i.e. $|\frac{\delta x}{x_0}| << 1\ \& \ |\frac{\delta y}{y_0}| << 1$. Now, writing them in the autonomous form explicitly, we get $ \frac{d}{dt}(x + \delta x) = f(x_0 + \delta x, y_0 + \delta y)$. Now, the total derivative of this with respect to time would look like 

\begin{equation}\dot{\delta x}  = f(x_0,y_0) + \big(\frac{\partial f}{\partial x} \big)_{(x_0,y_0)} \delta x + \big(\frac{\partial f}{\partial y} \big)_{(x_0,y_0)} \delta y \\ \dot{\delta y}  = g(x_0,y_0) + \big(\frac{\partial g}{\partial x} \big)_{(x_0,y_0)} \delta x + \big(\frac{\partial g}{\partial y} \big)_{(x_0,y_0)} \delta y \end{equation}

where LHS contains the derivative of these perturbations with respect to time.Since first derivatives $(f(x,y),\ g(x,y))$ are zero, the above set of equations takes the form 

\begin{equation}
\begin{bmatrix}
\dot{\delta x} \\ 
\dot{\delta y}  
\end{bmatrix}
= 
\begin{bmatrix}
\frac{\partial f}{\partial x} & \frac{\partial f}{\partial y} \\ 
\frac{\partial g}{\partial x} &  \frac{\partial g}{\partial y} 
\end{bmatrix} 
\begin{bmatrix}
\delta x \\ 
\delta y  
\end{bmatrix} 
\end{equation}
which is a linear system with respect to the perturbations $\delta x\ \& \ \delta y$.  Notice, these perturbations depend on both the magnitude of perturbations and spatial derivative of velocities linearly.

<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Now, the general solution of this system could be \begin{equation}X(t) = \alpha e^{\lambda_1 t} V_1 + \beta e^{\lambda_2 t} V_2 \end{equation}

where $\lambda_1 \ \& \ \lambda_2$ are eigenvalues and $V_1 \ \& \ V_2$ are eigenvectors of the matrix $A$ and $X(t)$ is $\begin{bmatrix}
{\delta x} \\ 
{\delta y}  
\end{bmatrix}$. 

We are equipped with techniques and a general formulation, so we'll do the ANALYSIS now. Here, we are not concerned with solving the equations precisely, but we want to know the general behaviour of solutions. Thus, one asks the question what happens if we start close to the equilibrium point, does it stay close or do something crazy?

Taking an example of a ball floating in a wave pool. And let's assume that I am studying the course of the ball in various different kind of ripples. But being a little lazy and a bit uninterested, I don't want to know where would this ball be after 5 minutes if I drop it somewhere in the pool. I am only interested in the qualitative behaviour of the ball. I am asking the questions like whether it will go crazy or it will be trapped in a whirlpool (yes, my wave pool can do that) or, rather uninterestingly, it will stay there only. These kinds of questions can be answered by looking at the critical points of differential equations. We can see the behaviour at critical points by looking at the eigenvectors and their corresponding eigenvalues. For that, we will take a look at the general solution we got and try to classify the different types of solutions for different eigenfunctions:

1. There are two exponential scalars in front of eigenvectors. These exponential terms will change their magnitude time changes. Thus, the eigenvectors will get either scaled up to infinity (when $\lambda s$ is positive) or scaled down to zero (when $\lambda s$ is negative) at $(x_0,\ y_0)$. 
2. The eigenvectors could be pointing parallel or perpendicularly. It gives diverse behaviour to the critical points. 

We will take a closer look to each of the case but before that, let's see how to plot these trajectories in python. 

## **Tips for writing the Code for Sketching Phase Portraits**

1. You can draw phase portraits for higher dimensions too. As an example of 2-D system, we discussed a little while, you can plot on the axes the spatial cordinates and over the trajectories indicate their velocity vectors with arrow. This is nothing new, you should have encountered [vector fields'](https://en.wikipedia.org/wiki/Vector_field#/media/File:VectorField.svg) pictorial representation.  
1. You need to draw trajectories as well as the velocity field. Therefore, you can simply plot trajectories (curves) by [streamplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.streamplot.html) and use [quiver plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.quiver.html) to plot velocity fields (arrows) over it. 
2. Remember, we are only be drawing trajectories originating at origin with zero velocities.  
3. Velocity field can be directly drawn from a given matrix. 
4. Further, matrix can be processed to obtain eigenvalues and eigenvectors which are to be plugged in general solution.   

Since our aim is to give you an overall guidance for both computing and physics, you are recommended to try out each function that we have used in this code. Though our code is highly commented, but reading about them from documentations is recommended.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Given some limit of phase space, system of equation and initial condtions the function should give you 
#################################### Phase Portraits ################################################### 


# defining the derivatives. Matrix multiplication 
def sys(Y):
    y1, y2 = Y
    return  A[0,0]*y1 + A[0,1]*y2, A[1,0]*y1 + A[1,1]*y2


def phase_portraits(f):
    
    # grid points for our spacial function quiver 
    y1 = np.linspace(-2*np.pi,2*np.pi , 50)
    y2 = np.linspace(-2*np.pi, 2*np.pi, 50)
    
    # defining grid here, it will give you ordered pairs 
    Y1, Y2 = np.meshgrid(y1, y2)

    
    # defining the empty array for derivatives
    u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)
    
    # From the function passed which is essentially discribing your system, 
    # We are going to find derivative at each point so as to find the velocities. 
    for i in range(np.size(y1)):
        for j in range(np.size(y2)):
            u[i,j],v[i,j]= f([Y1[i, j], Y2[i, j]])   
    
    # Here, function f needs to give you speeds in different directions as a function of its position
    # You can also simply give the matrix as parameter and manipulate it here to compute the velocities. 

    ############################################# Plotting ####################################################
    ###########################################################################################################        
    plt.figure()
    S = plt.streamplot(Y1, Y2, u, v,color='r')
    Q = plt.quiver(Y1, Y2, u, v,color='b')
    plt.xlabel('$y_1$')
    plt.ylabel('$y_2$')
    plt.scatter(0,0)
    plt.show()       
```

```python
# Since any system is recognized by the matrix A, we'll define here a random matrix.

A = np.matrix([[-.5, 1],[-1, -.5]])
phase_portraits(sys)
```
